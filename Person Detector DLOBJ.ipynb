```python
import os
import cv2

from imageai.Detection import VideoObjectDetection
from matplotlib import pyplot
from matplotlib import pyplot as plt
from matplotlib.patches import Rectangle
```

```python
# Will be used for storing each frame with detected object information.
processed_frame = []
```

```python
def forFrame(frame_number, output_array, output_count, returned_frame):
    """Function for bounding each detected person with a Red Box using BBox Coordinates"""
    plt.clf()

    this_colors = []
    labels = []
    sizes = []
    
    pyplot.imshow(returned_frame)
    ax = pyplot.gca()
    
    # Bounding each detected person object with Red Bounding Box.
    for box in output_array:
        ob_name = box["name"]                                               # name of the detected object
        ob_prob = box["percentage_probability"]                             # probability of detected object
        ob_cord = box["box_points"]                                         # box coordinates
        if ob_name == "person":
            x1 = ob_cord[0]
            y1 = ob_cord[1]
            x2 = ob_cord[2]
            y2 = ob_cord[3]
            width, height = x2 - x1, y2 - y1
            returned_frame = cv2.rectangle(returned_frame,(x1,y1),(x2,y2),(0,0,255),2)
    # Storing each frame with detected objects for future use [Video/Debugging etc]
    processed_frame.append(returned_frame)
```

```python
# Orginal Video Info
org_video_file_name = "Car - 16849.mp4" # Specify the video file name here
org_video_file_path = "../Original/"+org_video_file_name

# Generating Detected file Info
detected_file_name = org_video_file_name.split(".")[0] + "_detected"
detected_file_path = "../Detected/"+detected_file_name

# Fetching the FPS value
org_video = cv2.VideoCapture(org_video_file_path)
video_fps = int(org_video.get(cv2.CAP_PROP_FPS))
```

```python
"""
Model Details:
    - Yolo-V3
    - COCODATASET
    - frames_per_second = video_fps [Input Videos Frame Per Second].
    - return_detected_fraem = True  | Return each frme with detected objects.
    - minimum_percentage_probability - thresholding for defining integrity of detected objects.
    - per_frame_function = forFrame - used to specify function needs to be applied on each frame of the video.
"""
execution_path = os.getcwd()

# Initializing VideoDetector Object and Loading the pre trained yolo weights.
detector = VideoObjectDetection()
detector.setModelTypeAsYOLOv3()
detector.setModelPath( os.path.join(execution_path , "../Model/yolo.h5"))
detector.loadModel()


video_path = detector.detectObjectsFromVideo(input_file_path=org_video_file_path,save_detected_video = False,
                                             frames_per_second=video_fps, per_frame_function = forFrame,
                                             log_progress=True,
                                             return_detected_frame=True)
```

```python
# Defining output file name and fetching video dimensions.
output_file_name = detected_file_path+".avi"
height,width,layers = processed_frame[0].shape

# Defining videowriter object.
out = cv2.VideoWriter(output_file_name,cv2.VideoWriter_fourcc(*'MP4V'),video_fps,(width,height))

# Writing each processed frame to videowriter object and releasing the video.
[out.write(fname) for fname in processed_frame]
out.release()
```
